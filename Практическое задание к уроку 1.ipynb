Даны два вектора в трехмерном пространстве: (20,20,20) и (0,0,-20)

2) Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
def length_vector(a):
    result = 0
    for i in range(len(a)):
        result+= a[i]**2
    return np.sqrt(result)
```

```python
x1 = np.array([20,20,20])
x2 = np.array([0,0,-20])
```

```python
print(f'Длина вектора x1 равна {length_vector(x1)}')
print(f'Длина вектора x2 равна {length_vector(x2)}')
```

##### 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:

    1) окружности,
    
    2) эллипса,
    
    3) гиперболы.

```python
import math
from numpy import linalg
```

```python
def show_ellipse(a, b):
    if (a==0) or (b==0):
        print(f'Неверные параметры элипса{a,b}')
        return
    x = []
    y1 = []
    y2 = []
    for i in range(-1000*a, 1000*a+1):
        x_=i/1000
        x.append(x_)
        y1.append(math.sqrt(b**2 - (b**2) / (a**2) * x_**2))
        y2.append(-math.sqrt(b**2 - (b**2) / (a**2) * x_**2))
    plt.rcParams['axes.grid'] = True
    plt.axis('scaled')
    plt.plot(x,y1, color = 'black')
    plt.plot(x,y2, color = 'black')
    plt.xlabel('x')
    plt.xlabel('y')
    plt.ylim(-2*b-0.1,2*b+0.1)
    plt.xlim(-2*a-0.1,2*a+0.1)
```

```python
def show_circle(a):
    if (a==0):
        print(f'Неверные параметры окружности{a}')
        return
    show_ellipse(a,a)
```

```python
def show_hyperbola(a, b):
    if (a==0) or (b==0):
        print(f'Неверные параметры гиперболы{a,b}')
        return
    x1 = []
    x2 = []
    y1 = []
    y2 = []
    for i in range(1000*2*a):
        x_=i/1000
        if (-b**2 + (b**2) / (a**2) * x_**2) >= 0:
            x1.append(x_)
            x2.append(-x_)
            y1.append(math.sqrt(-b**2 + (b**2) / (a**2) * x_**2))
            y2.append(-math.sqrt(-b**2 + (b**2) / (a**2) * x_**2))

    plt.rcParams['axes.grid'] = True
    plt.axis('scaled')
    plt.plot(x1,y1,  color = 'black')
    plt.plot(x1,y2, color = 'black')
    plt.plot(x2,y1,  color = 'black')
    plt.plot(x2,y2, color = 'black')
    plt.xlabel('x')
    plt.xlabel('y')
    plt.ylim(-2*b-0.1,2*b+0.1)
    plt.xlim(-2*a-0.1,2*a+0.1)
```

```python
show_circle(2)
```

```python
show_ellipse(3,2)
```

```python
show_hyperbola(2,2)
```

1) Напишите код, который будет переводить полярные координаты в декартовы.


```python
def polar_to_dekart(R, alpha):
    x = R * np.cos(alpha)
    y = R * np.sin(alpha)
    return x,y
```

```python
polar_to_dekart(1, np.pi)
```
